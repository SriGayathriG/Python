```python
'''
step 1 : to plot time and heart rate
'''
plt.plot(cumulative_time_list,artifact_hr_list,linewidth=0.5)
plt.title('Heart rate graph')
plt.xlabel('time')
plt.ylabel('heart rate')
plt.show()
```

```python
'''
step 3 : plot the noise
'''
noise = noise.tolist() #to convert data elements of array into list
plt.plot(cumulative_time_list,noise,linewidth=0.5)
plt.title('Noise signal')
plt.xlabel('time')
plt.ylabel('noise')
plt.show()
```

```python
heart_rate_array = np.array(artifact_hr_list)
noise_array = np.array(noise)
print(heart_rate_array)
print(noise_array)
```

```python
'''
combining heart rate signal and noise
'''
combined_signal = heart_rate_array + noise_array
print(combined_signal)
plt.plot(cumulative_time_list,combined_signal,linewidth=0.5)
plt.title('Noise added signal')
plt.xlabel('time')
plt.ylabel('HR')
plt.show()
```

```python
'''
signal to noise ratio of heart rate only
'''
import numpy as np   
def signaltonoise(heart_rate_arr, axis=0, ddof=0):
    heart_rate_array = np.asanyarray(heart_rate_arr) 
    mean_of_signal = heart_rate_array.mean(axis)
    standard_deviation_signal = heart_rate_arr.std(axis=axis, ddof=ddof)
    return np.where(standard_deviation_signal == 0, 0, mean_of_signal/standard_deviation_signal)
print(signaltonoise(heart_rate_array,axis=0,ddof=0))
```

```python
'''
signal to noise ratio of noise only
'''
import numpy as np   
def signaltonoise(noise_array, axis=0, ddof=0):
    noise_array = np.asanyarray(noise_array)
    mean_noise = noise_array.mean(axis)
    standard_deviation_noise = noise_array.std(axis=axis, ddof=ddof)
    return np.where(standard_deviation_noise == 0, 0, mean_noise/standard_deviation_noise) 
print(signaltonoise(noise_array,axis=0,ddof=0))
```

```python
'''
signal to noise ratio of filtered signal
'''
import numpy as np   
def signaltonoise(filtered_signal, axis=0, ddof=0):
    filtered_signal = np.asanyarray(filtered_signal)
    mean_filtered_signal = filtered_signal.mean(axis)
    sd_filtered_signal = filtered_signal.std(axis=axis, ddof=ddof)
    return np.where(sd_filtered_signal == 0, 0, mean_filtered_signal/sd_filtered_signal) 
print(signaltonoise(filtered_signal,axis=0,ddof=0))
```
