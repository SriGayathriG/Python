```python
# https://www.codespeedy.com/calculate-signal-to-noise-ratio-in-python/
# https://github.com/hrtlacek/SNR/blob/main/SNR.ipynb

#signal to noise ratio of heart rate only
import numpy as np   
def signaltonoise(heart_rate_arr, axis=0, ddof=0):
    heart_rate_arr = np.asanyarray(heart_rate_arr) 
    me = heart_rate_arr.mean(axis)
    sd = heart_rate_arr.std(axis=axis, ddof=ddof)
    return np.where(sd == 0, 0, me/sd)
    # numpy.where(condition[, x, y])
    # https://note.nkmk.me/en/python-numpy-where/
print(signaltonoise(heart_rate_arr,axis=0,ddof=0))
```

```python
#signal to noise ratio of noise only
import numpy as np   
def signaltonoise(noise_arr, axis=0, ddof=0):
    noise_arr = np.asanyarray(noise_arr)
    me = noise_arr.mean(axis)
    sd = noise_arr.std(axis=axis, ddof=ddof)
    return np.where(sd == 0, 0, me/sd) 
print(signaltonoise(noise_arr,axis=0,ddof=0))
```

```python
#signal to noise ratio of both heart rate and noise 
import numpy as np   
def signaltonoise(final_signal, axis=0, ddof=0):
    final_signal = np.asanyarray(final_signal)
    me = final_signal.mean(axis)
    sd = final_signal.std(axis=axis, ddof=ddof) #degree of freedom
    return np.where(sd == 0, 0, me/sd)
print(signaltonoise(final_signal,axis=0,ddof=0))
```
