```python
labels = ['green','blue','red','black']
bins = [20,30,40,85,100]
df['%VO2max'] = pd.cut(df['%VO2max'],bins = bins, labels = labels, right=False)
```

```python
df.head()
```

```python
xyz = df['%VO2max']
xyz = list(xyz)
zzz = []
```

```python
for i in xyz:
  if (0 <= i <= 20):
    zzz.append('green')
  elif (21 <= i <= 30):
    zzz.append('blue')
  elif (31 <= i <= 40):
    zzz.append('red')
  elif (i >= 41):
    zzz.append('black')
```
